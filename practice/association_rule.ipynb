# Step0. 모듈 불러오기

```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

# Step1. 분석에 필요한 데이터 구성하기

```python
ds_asso = pd.read_csv("상품구매.csv",engine = "python")
ds_asso
```

```python
#ds_asso.groupby('ID').groups
lst = []
for name, group in ds_asso.groupby("ID"):
    lst.append(group["PRODUCT"].tolist())
```

```python
enc = TransactionEncoder()
arr = enc.fit_transform(X = lst)
ds_pre = pd.DataFrame(arr, columns=enc.columns_)
ds_pre
```

# Step2. 라이브러리 불러오기

### 최소 지지도 0.4로 설정

```python
min_support = 0.4
ds_freq = apriori(ds_pre, min_support= min_support, use_colnames=True)
ds_freq.round(3)
```

```python
ds_asso_rule = association_rules(ds_freq, metric = "confidence", min_threshold= 0.4)
ds_asso_rule.round(3)
```
